```python
img = cv2.imread(img_path)

if img is None:
    print("Image not loaded. Path is wrong.")
else:
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
x_min = 120
y_min = 80
x_max = 350
y_max - 460
label = "Disease"

cv2.rectangle(img, (x_min, y_min), (x_max, y_max), (255, 0, 0), 2)
cv2.putText(img, label, (x-min, y_min - 5),
            cv2.FONT_HERSHEY_SIMPLEX, 0.7,(255, 0, 0), 2)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis("off")
plt.show()
```
